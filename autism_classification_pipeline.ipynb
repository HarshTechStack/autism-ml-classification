```python
# LABEL ENCODING
from sklearn.preprocessing import LabelEncoder
df1 = df
cat =['gender', 'jundice', 'austim', 'used_app_before','Class/ASD']
le = LabelEncoder()
df1[cat]=df1[cat].apply(le.fit_transform)
```

# Visualisation

```python
import matplotlib.pyplot as plt

# Calculate the counts of each ASD trait category
asd_counts = df['Class/ASD'].value_counts()

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the pie chart
ax.pie(asd_counts, labels=asd_counts.index, autopct='%1.1f%%')

# Set the title
ax.set_title('Autism Spectrum Disorder Counts', fontsize=20)

# Show the chart
plt.show()
```

```python
Y =  df1[['Class/ASD']]  # target
X = df1.drop(columns=['Class/ASD'])  # feature

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
len(x_train),len(x_test),len(y_train),len(y_test)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rfmodel = RandomForestClassifier()
rfmodel.fit(x_train, y_train)
y_pred1 = rfmodel.predict(x_test)
y_pred1
```

```python
from sklearn.model_selection import cross_val_score
accuracies= cross_val_score(estimator=rfmodel,X=X,y=Y,cv=5)
print(accuracies*100)
print(accuracies.mean()*100)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
print("ACCURACY : ",accuracy_score(y_test,y_pred1))
```

```python
accuracy_RF = accuracy_score(y_test, y_pred1)
accuracy_RF = np.round(accuracy_RF,3)
print("Accuracy is", accuracy_RF*100,"%")
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

# Choose the number of neighbors (k)
k = 5
kncmodel = KNeighborsClassifier(n_neighbors=k)
kncmodel.fit(x_train, y_train)
y_pred2 = kncmodel.predict(x_test)
y_pred2
```

```python
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
accuracy_KNN = accuracy_score(y_test, y_pred2)
accuracy_KNN = np.round(accuracy_KNN,3)
print("Accuracy is", accuracy_KNN*100,"%")
```

# AUC - ROC Curve

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
ns_pred = [0 for _ in range(len(y_test))]  #no skillÂ prediction
```

```python
ns_auc = roc_auc_score(y_test,ns_pred)
rf_auc = roc_auc_score(y_test,y_pred1)
knn_auc = roc_auc_score(y_test,y_pred2)
xg_auc = roc_auc_score(y_test,y_pred3)
cat_auc = roc_auc_score(y_test,y_pred4)
```

```python
print('No-Skill Prediction: AUROC = %.3f' % (ns_auc))
print('Random Forest: AUROC = %.3f' % (rf_auc))
print('kNN: AUROC = %.3f' % (knn_auc))
print('XG Boost: AUROC = %.3f' % (xg_auc))
print('Cat Boost: AUROC = %.3f' % (cat_auc))
```

```python
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_pred)
rf_fpr, rf_tpr, _ = roc_curve(y_test, y_pred2)
knn_fpr, knn_tpr, _ = roc_curve(y_test, y_pred4)
xg_fpr, xg_tpr, _ = roc_curve(y_test, y_pred3)
cat_fpr, cat_tpr, _ = roc_curve(y_test, y_pred1)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No skill prediction (AUC = %0.3f)' % ns_auc)
plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUC = %0.3f)' % rf_auc)
plt.plot(knn_fpr, knn_tpr, marker='.', label='kNN (AUC = %0.3f)' % knn_auc)
plt.plot(xg_fpr, xg_tpr, marker='.', label='XG Boost (AUC = %0.3f)' % xg_auc)
plt.plot(cat_fpr, cat_tpr, marker='.', label='Cat Boost (AUC = %0.3f)' % cat_auc)

plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
